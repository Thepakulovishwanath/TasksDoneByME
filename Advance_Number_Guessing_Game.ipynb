```python
import random

def get_hint(guess_num, random_Number):
    """
    Returns a hint based on how close the guess is to the actual number
    """
    if abs(guess_num - random_Number) <= 10:
        return "You're getting warmer!"
    elif abs(guess_num - random_Number) <= 20:
        return "You're getting close!"
    elif abs(guess_num - random_Number) <= 30:
        return "You're on the right track!"
    else:
        return "Too cold!"

def play_game():
    """
    Main loop to play the guessing game
    """
    while True:
        # Initialize the number of attempts
        attempts = 0
        
        # Generate a random number between 1 and 100
        random_Number = random.randint(1, 100)
        
        # Loop until the user guesses the correct number
        while True:
            # Increment the number of attempts
            attempts += 1
            
            # Prompt the user to enter a guess
            guess_num = input("Enter a Number to Guess (Between 1 To 100): ")
            
            # Check if the input is a valid digit
            if guess_num.isdigit():
                # Convert the input to an integer
                guess_num = int(guess_num)
            else:
                # Handle invalid input (non-numeric characters)
                print("Enter a valid digit.")
                continue
            
            # Check if the guess is correct
            if random_Number == guess_num:
                print("Your guess is correct!")
                break
            # Check if the guess is too low
            elif random_Number > guess_num:
                print("Your guess is too low.")
                print(get_hint(guess_num, random_Number))
            # Check if the guess is too high
            else:
                print("Your guess is too high.")
                print(get_hint(guess_num, random_Number))
        
        # Display the number of attempts taken to guess the correct number
        print("You got it in", attempts, "guesses!")
        
        # Ask the user if they want to continue playing
        result = input("Do you want to continue? Press 1 for Yes, 0 for No: ")
        
        # Check if the user wants to continue
        if result == '1':
            # Continue to the next iteration of the main loop
            continue
        else:
            # Exit the game
            print("Thank you for playing!")
            break

# Call the play_game function to start the game
play_game()
```
